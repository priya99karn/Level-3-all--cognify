# TASK 2

# <span style="color:PURPLE">Cuisine Combination</span>

### Identify the most common combinations of
### cuisines in the dataset.
### Determine if certain cuisine combinations
### tend to have higher ratings.

```python
from collections import Counter

# Split the cuisines into lists and count the combinations
df['Cuisines List'] = df['Cuisines'].astype(str).apply(lambda x: tuple(sorted(x.split(', '))))
cuisine_combinations = Counter(df['Cuisines List'])

# Find the most common cuisine combinations
most_common_cuisines = cuisine_combinations.most_common(10)

most_common_cuisines
```

```python
# Calculate the average aggregate rating for each cuisine combination
average_ratings_by_cuisine = df.groupby('Cuisines List')['Aggregate rating'].mean().sort_values(ascending=False)

# Display the top 10 highest rated cuisine combinations
highest_rated_cuisines = average_ratings_by_cuisine.head(10)
highest_rated_cuisines

```

```python
#graphical represntation of higjes rated cuisine combination
```

```python
# Convert the Series to DataFrame for plotting
highest_rated_cuisines_df = highest_rated_cuisines.reset_index()
highest_rated_cuisines_df['Cuisines Combined'] = highest_rated_cuisines_df['Cuisines List'].apply(lambda x: ', '.join(x))

# Plotting
plt.figure(figsize=(14, 8))
sns.barplot(x='Aggregate rating', y='Cuisines Combined', data=highest_rated_cuisines_df, palette="viridis")
plt.title('Top 10 Highest Rated Cuisine Combinations')
plt.xlabel('Average Aggregate Rating')
plt.ylabel('Cuisine Combinations')
plt.xlim(4.5, 5)  # Adjusting the x-axis to better visualize the differences in ratings
plt.show()
```

# TASK 4

# <span style="color:PURPLE">Restaurant Chains</span>
### Identify if there are any restaurant chains
### present in the dataset.
### Analyze the ratings and popularity of
### different restaurant chains

```python
# identifying rerstaurant chains
```

```python
# Count occurrences of each restaurant name
restaurant_counts = df['Restaurant Name'].value_counts()

# Filter out names that occur more than once to identify chains
restaurant_chains = restaurant_counts[restaurant_counts > 1]

# Display the top 10 restaurant chains by number of locations
restaurant_chains.head(10)
```

```python
# analysing ratings

# Calculate the average aggregate rating for each restaurant chain
average_ratings = df[df['Restaurant Name'].isin(restaurant_chains.index)].groupby('Restaurant Name')['Aggregate rating'].mean()
average_ratings
```

```python
# analysisng popularity
```

```python
# Calculate the average number of votes for each restaurant chain
average_votes = df[df['Restaurant Name'].isin(restaurant_chains.index)].groupby('Restaurant Name')['Votes'].mean()
average_votes
```

```python
# visualizing top restautrant chains using visualization
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Top 10 restaurant chains by number of outlets
top_chains = restaurant_chains.head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_chains.values, y=top_chains.index, palette='viridis')
plt.title('Top 10 Restaurant Chains by Number of Outlets')
plt.xlabel('Number of Outlets')
plt.ylabel('Restaurant Chain')
plt.show()
```

```python
# Visualize Average Ratings of Top Restaurant Chains

# Top 10 restaurant chains by average rating
top_ratings = average_ratings.nlargest(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_ratings.values, y=top_ratings.index, palette='coolwarm')
plt.title('Average Ratings of Top Restaurant Chains')
plt.xlabel('Average Rating')
plt.ylabel('Restaurant Chain')
plt.show()
```

```python
# Visualize Average Popularity (Votes) of Top Restaurant Chains

# Top 10 restaurant chains by average votes
top_votes = average_votes.nlargest(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_votes.values, y=top_votes.index, palette='mako')
plt.title('Average Popularity (Votes) of Top Restaurant Chains')
plt.xlabel('Average Votes')
plt.ylabel('Restaurant Chain')
plt.show()
```
